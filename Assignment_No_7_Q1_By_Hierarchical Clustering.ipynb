# Hierarchical Clustering :-

```python
# Normalize heterogenous numerical data
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot as plt
from sklearn.preprocessing import normalize
```

```python
airline1_norm = pd.DataFrame(normalize(airline1),columns=airline1.columns)
airline1_norm.head()
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(airline1_norm,'complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage='ward')
hc
```

```python
y = pd.DataFrame(hc.fit_predict(airline1_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
airline1['clustersid'] = hc.labels_
airline1
```

```python
# Ploting Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airline1['clustersid'],airline1['Balance'], c=hc.labels_) 
```
